```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = lambda x: x - (np.log(x))**x
b = lambda x: np.log(x) - x**2 + 2.5*x - 1

n = 100

x = np.linspace(4, 5, n)
plt.plot(x,a(x))
plt.hlines(0,4,5,'black')
plt.grid(True)
plt.show()

x = np.linspace(0.5, 1, n)
plt.plot(x,b(x))
plt.hlines(0,0.5,1,'black')
plt.grid()
plt.show()
```

```python
a = lambda x: x - 3**-x
b = lambda x: x**3 - 2*x**2 -4*x + 3

n = 100

li, ls = -1, 3
x = np.linspace(li, ls, n)
plt.plot(x,a(x))
plt.hlines(0, li, ls, 'black')
plt.grid(True)
plt.show()

li, ls = -2, 4
x = np.linspace(li, ls, n)
plt.plot(x,b(x))
plt.hlines(0, li, ls, 'black')
plt.grid(True)
plt.show()
```

```python
a = lambda x: (x-1)*np.tan(x) + x*np.sin(np.pi*x)
b = lambda x: (x-2)*np.sin(x)*np.log(x+2)

n = 100

li, ls = 0, 1
x = np.linspace(li, ls, n)
plt.plot(x,a(x))
plt.hlines(0, li, ls, 'black')
plt.grid(True)
plt.show()

li, ls = -1, 3
x = np.linspace(li, ls, n)
plt.plot(x,b(x))
plt.hlines(0, li, ls, 'black')
plt.grid(True)
plt.show()
```

```python
a = lambda x: abs((2-np.exp(x)+2*x)/3)
b = lambda x: abs((4*x-3)/(x**2-2*x))

n = 100

li, ls = 0, 1
x = np.linspace(li, ls, n)
plt.plot(x,a(x))
plt.hlines(0, li, ls, 'black')
plt.grid(True)
plt.show()

li, ls = 0.5, 1
x = np.linspace(li, ls, n)
plt.plot(x,b(x))
plt.hlines(0, li, ls, 'black')
plt.grid(True)
plt.show()
```
