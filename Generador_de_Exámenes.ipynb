```python
from random import randint
import pylatex as plx
from pylatex.package import Package
from pylatex.utils import bold, NoEscape
from pylatex.base_classes import Environment, SpecialOptions, Arguments
from pylatex.basic import NewLine
```

```python
def loteria_preguntas(preguntas, np):
  n = len(preguntas)
  lp = []
  if n < np:
    print('Hay menos preguntas de las que se desean seleccionar.')
  else:
    p = preguntas.copy()
    for i in range(np):
      pregunta = p[randint(0,n - 1 - i)]
      lp.append(pregunta)
      p.remove(pregunta)
  return lp
```

```python
def selector_preguntas(texto, np):
  preguntas = tex.split('\\question')
  lista = []
  for p in preguntas:
    if p[0] == '[':
      p = p.split('\n')
      lista.append(p[0])

  return loteria_preguntas(lista, np)
```

```python
class tabular_raro(Environment):
  _latex_name='tabular*'
```

```python
class center(Environment):
  _latex_name='center'
```

```python
class quest(Environment):
  _latex_name='questions'
```

```python
def generador(filename='Examen', clas='Métodos Numéricos I', exam_name='Primer Examen Parcial', date = '23 de octubre de 2024', time = '80 minutos', questions= ['¿Hola como está?']):
  doc = plx.Document(filename,
                    documentclass=plx.Command('documentclass', options=("a4paper", "10pt", "legalpaper"), arguments='exam'),
                    lmodern=False, textcomp=False)

  doc.packages.append(Package('babel', 'spanish'))
  doc.packages.append(Package('geometry', 'margin=.75in'))
  doc.packages.append(Package('amsmath,amssymb'))
  doc.packages.append(Package('multicol'))
  doc.packages.append(Package('graphicx'))
  doc.packages.append(Package('tikz'))

  doc.preamble.append(plx.Command('pagestyle',arguments='head'))
  doc.preamble.append(plx.Command('firstpageheader',arguments=('Escuela Superior de Física y Matemáticas','','Instituto Politécnico Nacional')))
  doc.preamble.append(plx.Command('runningheadrule'))

  doc.preamble.append(plx.Command('pointpoints', arguments=('punto', 'puntos')))
  doc.preamble.append(plx.Command('hpword', arguments='Puntos:'))
  doc.preamble.append(plx.Command('vpword', arguments='Puntos'))
  doc.preamble.append(plx.Command('vtword', arguments='Total:'))
  doc.preamble.append(plx.Command('htword', arguments='Total'))
  doc.preamble.append(plx.Command('vsword', arguments='Resultado'))
  doc.preamble.append(plx.Command('hsword', arguments='Resultado:'))
  doc.preamble.append(plx.Command('vqword', arguments='Problema'))
  doc.preamble.append(plx.Command('hqword', arguments='Pregunta:'))


  doc.append(plx.Command('noindent'))

  with doc.create(tabular_raro(arguments=[plx.Command('textwidth'),NoEscape('l @{\\extracolsep{\\fill}} r @{\\extracolsep{6pt}} l')])) as tab:
    tab.append(NoEscape(bold(clas) +'&'+ bold('Nombre:') +'&'+ plx.Command('makebox', options='2.5in', arguments=plx.Command('hrulefill')).dumps()+'\\\\'))
    d = bold('Calificación:') + plx.Command('makebox', options='0.5in', arguments=plx.Command('hrulefill')).dumps()
    tab.append(NoEscape(bold(exam_name) +'&'+ plx.MultiColumn(2,align='r',data=NoEscape(d)).dumps()+'\\\\'))
    tab.append(NoEscape(bold('Fecha:'+date) +'&'+ plx.MultiColumn(2,align='r',data=bold('Tiempo:'+time)).dumps()+'\\\\'))

  doc.append(NewLine())
  doc.append(plx.Command('rule',options='2ex',arguments=(NoEscape('\\textwidth'), '2pt')))
  doc.append(NewLine())


  doc.append(plx.Command('noindent'))

  doc.append(NoEscape('Este examen contiene \\numquestions \\ ejercicios que corresponden a \\numpoints \\ puntos de la primera calificación parcial del curso. Para la calificación de cada ejercicio se consideran los siguientes valores: Planteamineto correcto: 25 \\%; Desarrollo correcto: 50 \\%; Solución correcta (incluye uso correcto de la aproximación): 25 \\%'))


  with doc.create(center()) as c:
    c.append('Tabla de calificación de uso exclusivo para el profesor.')
    c.append(plx.Command('par'))
    c.append(plx.Command('addpoints'))
    c.append(plx.Command('gradetable', options= SpecialOptions(['h', 'questions'])))

  with doc.create(quest()) as q:
    for p in questions:
      q.append(plx.Command('addpoints'))
      q.append(plx.Command('question', options='05'))
      q.append(NoEscape(p))

  doc.append(plx.Command('noindent'))
  doc.append(plx.Command('rule',options='2ex',arguments=(NoEscape('\\textwidth'), '2pt')))
  doc.append(NewLine())

  with doc.create(center()) as c:
    c.append(bold('Solución'))



  doc.generate_tex()

```

```python
f = open("document.tex", "r")
tex = f.read()
```

```python
p = selector_preguntas(tex,2)
print(p)
```

```python
generador(questions=p)
```
